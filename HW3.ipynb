```python
import matplotlib.pyplot as plt
import numpy as np
```

### Example 1-1
基本np.array與matplotlib繪圖

```python
A = np.array([1,2,3,4])
B = np.array([1,2,3,4])
print(A+B)
print(A*B)
plt.plot(A, B, '-o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Example 1-2
基本np.array與matplotlib繪圖

```python
xArr = np.linspace(0, 2*np.pi, 20)
yArr = np.sin(xArr)
print(xArr)
print(yArr)
plt.plot(xArr, yArr, '-o', label='sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

### Example 3

```python
a1 = np.array([1,0])
a2 = np.array([0,1])
atom1 = np.array([0.0, 0.0])
atom2 = np.array([0.5, 0.5])

# plot a1 and a2
plt.arrow(0, 0, a1[0], a1[1], head_width=0.15, head_length=0.15, fc='k', ec='k')
plt.arrow(0, 0, a2[0], a2[1], head_width=0.15, head_length=0.15, fc='k', ec='k')

# plot atoms
for i in range(4):
    for j in range(4):
        atom1_new = atom1 + a1*i + a2*j
        atom2_new = atom2 + a1*i + a2*j
        plt.plot(atom1_new[0], atom1_new[1], 'or')
        plt.plot(atom2_new[0], atom2_new[1], 'ob')

plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
